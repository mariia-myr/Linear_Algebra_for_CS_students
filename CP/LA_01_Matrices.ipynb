```python
# these imports may be useful

import pandas as pd
import numpy as np
```

 #### –î–µ—è–∫—ñ —ñ–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–∏, —è–∫—ñ, –º–æ–∂–ª–∏–≤–æ, —Å—Ç–∞–Ω—É—Ç—å —É –Ω–∞–≥–æ–¥—ñ. 

```python
# these tools may be useful
# give a matrix
a = np.array([[1,2], [2,4], [3,6]]) #3x2 matrix
b = np.array([[1], [2]])            #2x1 matrix = column vector

#zero matrix
z = np.zeros((2, 3))   

#matrix with all 1 elements
u = z + 1

# transpose
at = a.transpose()

# matrix multiplication
prod1 = a @ b
prod2 = np.matmul(a, b)

# dot (inner, scalar) product of vectors
u = np.array([1,2,3])
v = np.array([2,3,4])
uv = np.dot(u, v)

print(f' zero={z},\n all 1-matrix={u},\n a= {a},\n b = {b},\n a^t = {at},\n ab = {prod1}, \n ab= {prod2}')
print(f'u={u}, v={v}, uv = {uv}')
```
